**Now using matplpotlib,seaborn to plot some visuals for realtionship**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

titanic = sns.load_dataset('titanic')
titanic.head()
```

#joint plot

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

**Box plot**

```python
sns.boxplot(x='pclass',y='age',data=titanic)
```

```python
sns.swarmplot(x='alive',y='age',data=titanic)
```

```python
sns.countplot(x='sex',data=titanic)
```

**we can use dashes parameter to remove dotted line by giving it false OR "In seaborn legend() function is used to remove legend of plot but if not used then it has always legend on graph"**

**now we used grid ot title parameter**

```python
sns.lineplot(x="pclass",y="age",
             data=titanic,hue="sex",style="sex",palette="plasma",
             markers=["o",">"])
plt.title('Titanic',fontsize='18')
plt.show()
```

# histogram plot method in seaborn

```python
sns.displot(titanic["age"])
plt.show()
```

# PLOTTING BAR CHART IN SEABORN

```python
sns.barplot(x="embark_town",y="age",data=titanic,
            hue="sex",hue_order=["female","male"]
            )
plt.show()
```

# HEATMAP METHOD OF PLOTTING IN SEABORN

```python
df=sns.load_dataset("penguins").head(20)
df
df=np.linspace(1,10,20).reshape(4,5)
df
sns.heatmap(df)
plt.show()
```

**some parameters as shown**

**annot parameter used to show the value on heatmap**

```python
df=np.linspace(1,10,10).reshape(2,5)
df
sns.heatmap(df,vmin=0,vmax=20,cmap="gist_gray",annot=True)
plt.show()
```
